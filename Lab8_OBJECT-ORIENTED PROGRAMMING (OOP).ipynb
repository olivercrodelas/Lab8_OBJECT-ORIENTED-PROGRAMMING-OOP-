```python
# 2. Implementing Encapsulation in Python

# Step 1: Define a class with private attributes
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self._balance = balance # Private attribute
    def deposit(self, amount):
        self._balance += amount
        return f"New balance: ${self._balance}"
    def get_balance(self):
        return f"Balance: ${self._balance}"

# Step 2: Create an object and test encapsulation
account = BankAccount("Alice", 5000)

# Step 3: Deposit money and display the updated balance
print(account.deposit(1500))
print(account.get_balance())


# --- Try It ---
class StudentRecord:
    def __init__(self):
        self._grades = [] # Private attribute
    def add_grade(self, grade):
        self._grades.append(grade)
    def retrieve_grades(self):
        return f"Student Grades: {self._grades}"

record = StudentRecord()
record.add_grade(95)
record.add_grade(88)
print(record.retrieve_grades())
```

```python
# 3. Applying Inheritance in Python

# Step 1: Define a base class
class Person:
    def __init__(self, name):
        self.name = name
    def introduce(self):
        return f"Hi, I'm {self.name}."

# Step 2: Define a subclass that inherits from Person
class Student(Person):
    def introduce(self):
        return f"Hi, I'm {self.name}, and I'm a student."

# Step 3: Create objects and call methods
person = Person("Allen")
student = Student("Mark")
print(person.introduce())
print(student.introduce())


# --- Try It ---
class Device:
    def __init__(self, device_name):
        self.device_name = device_name
    def power_on(self):
        return "Device is turning on"

class Smartphone(Device):
    def power_on(self):
        return f"{self.device_name} is booting up... Apple Logo appears."

generic_device = Device("Generic Gadget")
my_phone = Smartphone("iPhone 15")
print(generic_device.power_on())
print(my_phone.power_on())
```

```python
# 4. Utilizing Polymorphism in Python

# Step 1: Define a base class
class Shape:
    def area(self):
        return "Calculating area..."

# Step 2: Define subclasses with their implementations
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

# Step 3: Create objects and call their methods
circle = Circle(5)
rectangle = Rectangle(4, 6)
print(f"Circle area: {circle.area()}")
print(f"Rectangle area: {rectangle.area()}")


# --- Try It ---
class BaseDevice:
    def power_on(self):
        pass

class SmartPhone(BaseDevice):
    def power_on(self):
        return "Smartphone turning on..."

class Laptop(BaseDevice):
    def power_on(self):
        return "Laptop booting Windows..."

def start_device(device):
    print(device.power_on())

phone = SmartPhone()
laptop_device = Laptop()

start_device(phone)
start_device(laptop_device)
```

```python
# 6. Working with Class and Static Methods

# Step 1: Define a class with a static method
class MathOperations:
    @staticmethod
    def square(num):
        return num * num

# Step 2: Get user input
user_input = 5

# Step 3: Call the static method with user input
result = MathOperations.square(user_input)

# Step 4: Print the result
print("Square of", user_input, ":", result)


# --- Try It ---
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

celsius_input = 28
converted = TemperatureConverter.celsius_to_fahrenheit(celsius_input)
print(f"{celsius_input}°C is equal to {converted}°F")
```

```python
# 7. Operator Overloading in Python

# Step 1: Define a class with overloaded operators
class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    def __add__(self, other):
        return BankAccount(self.balance + other.balance)

# Step 2: Create account objects and add their balances
account1 = BankAccount(5000)
account2 = BankAccount(3000)
total_balance = account1 + account2 # Combines the balances

# Step 3: Print the result
print(f"Total Balance: P{total_balance.balance}")


# --- Try It ---
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(5, 10)
v2 = Vector(2, 3)
print(f"Vector Addition: {v1} + {v2} = {v1 + v2}")
print(f"Vector Subtraction: {v1} - {v2} = {v1 - v2}")
```

```python
# 8. Abstract Classes and Methods in Python
from abc import ABC, abstractmethod

# Step 1: Define an abstract class
class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

# Step 2: Define a subclass implementing the abstract method
class WashingMachine(Appliance):
    def turn_on(self):
        return "Washing machine is now running."

# Step 3: Create an object of the subclass and call the method
wm = WashingMachine()
print(wm.turn_on())


# --- Try It ---
class Refrigerator(Appliance):
    def turn_on(self):
        return "Refrigerator is now cooling."

fridge = Refrigerator()
print(fridge.turn_on())
```

```python
# 9. Implementing Method Resolution Order (MRO)

# Step 1: Define multiple classes
class A:
    def show(self):
        return "Class A"
class B(A):
    def show(self):
        return "Class B"
class C(A):
    def show(self):
        return "Class C"
class D(B,C):
    pass # Inherits show() based on MRO

# Step 2: Create an object of class D and call the method
obj = D()
print(obj.show()) # Output depends on MRO


# --- Try It ---
class BaseClass:
    def show(self): return "Base"
class ClassOne(BaseClass):
    def show(self): return "One"
class ClassTwo(BaseClass):
    def show(self): return "Two"
class ClassFinal(ClassOne, ClassTwo):
    pass 

print("MRO for ClassFinal:")
print(ClassFinal.mro())
```

```python
# --- 1. Library Book Management System ---
class Library:
    def __init__(self): self.books = {}
        
    def add_book(self, title, author):
        if type(title) is not str: return "Error: Title must be a string."
        self.books[title] = {'author': author, 'avail': True}
        return f"Book '{title}' added to the system."
        
    def borrow_book(self, title):
        if title not in self.books: return "Error: Book does not exist."
        if not self.books[title]['avail']: return "Error: Book is already borrowed."
        self.books[title]['avail'] = False; return f"Successfully borrowed '{title}'."
        
    def return_book(self, title):
        if title in self.books and not self.books[title]['avail']:
            self.books[title]['avail'] = True; return f"Successfully returned '{title}'."
        return "Error: Invalid return."
        
    def display_books(self):
        print("\n--- Library Books ---")
        for t, i in self.books.items():
            print(f"Title: {t} | Author: {i['author']} | Status: {'Available' if i['avail'] else 'Borrowed'}")

# --- 2. Employee Payroll System ---
class Payroll:
    def __init__(self): self.emps = []
        
    def add_employee(self, name, rate, hours):
        if rate < 0 or hours < 0: return "Error: Hours and wages cannot be negative."
        self.emps.append({'name': name, 'rate': rate, 'hours': hours}); return f"Employee {name} added."
        
    def calculate_payroll(self):
        if not self.emps: return "Error: At least one employee must be entered."
        print("\n--- Payroll Report ---")
        for e in self.emps:
            b_hrs, o_hrs = min(e['hours'], 160), max(0, e['hours'] - 160)
            print(f"Name: {e['name']} | Total Hours: {e['hours']} | Base Salary: ${b_hrs * e['rate']:.2f} | Overtime Pay: ${o_hrs * e['rate'] * 1.5:.2f}")

# --- 3. Online Shopping Cart System ---
class ShoppingCart:
    def __init__(self):
        self.inv = {"Laptop": {"p": 1000, "s": 5}, "Mouse": {"p": 25, "s": 10}, "Keyboard": {"p": 50, "s": 0}}
        self.cart = {}
        
    def add_item(self, item, qty):
        if item not in self.inv: return f"Error: '{item}' is not a valid product."
        cur_qty = self.cart.get(item, 0)
        if cur_qty + qty > self.inv[item]['s']: return f"Error: Cannot add {qty}x {item}. Only {self.inv[item]['s'] - cur_qty} left in stock."
        self.cart[item] = cur_qty + qty; return f"Successfully added {qty}x {item} to cart."
        
    def remove_item(self, item, qty):
        if item in self.cart:
            self.cart[item] = max(0, self.cart[item] - qty)
            if self.cart[item] == 0: del self.cart[item]
            return f"Successfully removed {qty}x {item} from cart."
        return f"Error: '{item}' is not in your cart."
        
    def view_cart(self):
        print("\n--- Shopping Cart ---")
        total = sum(self.inv[i]['p'] * q for i, q in self.cart.items())
        for i, q in self.cart.items(): print(f"{q}x {i} = ${self.inv[i]['p'] * q}")
        print(f"Total Price: ${total}")


# ==========================================
# TEST RUN OUTPUTS
# ==========================================

print("=== EXERCISE 1 TEST ===")
lib = Library()
print(lib.add_book("1984", "George Orwell"))
print(lib.borrow_book("1984"))
print(lib.borrow_book("1984")) 
lib.display_books()

print("\n=== EXERCISE 2 TEST ===")
payroll = Payroll()
print(payroll.add_employee("Alice", 20, 150)) 
print(payroll.add_employee("Bob", 25, 170))   
print(payroll.add_employee("Charlie", -10, 40)) 
payroll.calculate_payroll()

print("\n=== EXERCISE 3 TEST ===")
cart = ShoppingCart()
print(cart.add_item("Laptop", 1))
print(cart.add_item("Mouse", 12)) 
print(cart.add_item("Mouse", 2))
print(cart.remove_item("Mouse", 1))
cart.view_cart()
```
